```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class IntentPredictor(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(IntentPredictor, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Task types and modes
tasks = ['Sprint Planning', 'Bug Triage', 'Retrospectives', 'Brainstorming', 'Backlog Grooming']
modes = ['Focus', 'Ideation', 'Execution']

# Simulated AI trigger rates (0â€“100 scale)
data = np.array([
    [20, 70, 90],  # Sprint Planning
    [10, 30, 60],  # Bug Triage
    [15, 10, 5],   # Retrospectives
    [25, 80, 40],  # Brainstorming
    [30, 50, 70]   # Backlog Grooming
])
```

```python
# Data Visualization

fig, ax = plt.subplots(figsize=(8, 6))
heatmap = ax.imshow(data, cmap='YlOrRd')

# Add labels
ax.set_xticks(np.arange(len(modes)))
ax.set_yticks(np.arange(len(tasks)))
ax.set_xticklabels(modes)
ax.set_yticklabels(tasks)

# Rotate x-axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Annotate each cell
for i in range(len(tasks)):
    for j in range(len(modes)):
        text = ax.text(j, i, data[i, j], ha="center", va="center", color="black")

# Add title and colorbar
ax.set_title("AI Trigger Frequency by Task Type and Mode")
fig.colorbar(heatmap)
plt.tight_layout()
plt.show()
```
